```python
#handling null values

df.fillna(df.mean(), inplace=True) 
df.dropna(inplace=True)
```

```python
#scaling

scaler = StandardScaler()
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
df
```
